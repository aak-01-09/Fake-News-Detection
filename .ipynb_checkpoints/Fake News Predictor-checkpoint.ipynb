```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake = pd.read_csv('Fake.csv')
data_true = pd.read_csv('True.csv')
```

```python
data_fake["class"] = 0
data_true['class'] = 1
```

```python
data_fake_manual_testing = data_fake.tail(10)
for i in range (23480, 23470, -1):
    data_fake.drop([i], axis = 0, inplace = True)

data_true_manual_testing = data_true.tail(10)
for i in range (21416, 21406, -1):
    data_true.drop([i], axis = 0, inplace = True)
```

```python
data_merge = pd.concat([data_fake, data_true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title', 'subject', 'date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.columns 
```

```python
data.head()
```

```python
print(data['text'].head(10))         # see actual first rows
print(data['text'].isnull().sum())   # check for missing values
print(data['text'].apply(lambda x: len(str(x).strip())).head(10))  # length of text
```

```python
data = data[data['text'].apply(lambda x: len(str(x).strip()) > 0)]
```

```python
def wordopt(text):
    text = str(text).lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # remove URLs
    text = re.sub(r'<.*?>', '', text)                  # remove HTML
    text = re.sub(r'\[.*?\]', '', text)                # remove brackets
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)          # keep letters + spaces
    text = re.sub(r'\s+', ' ', text)                  # normalize spaces
    return text.strip()

data['text'] = data['text'].apply(wordopt)
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
print("Empty rows now:", (data['text'].str.strip() == "").sum())
print(data['text'].head())
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
print("Unique predictions:", set(pred_lr))
print("Actual classes:", set(y_test))
```

```python
def output_label(n):
    if n == 0:
        return"Fake News"
    elif n == 1:
        return "Not a Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GBC = GB.predict(new_xv_test)
    pred_RFC = RF.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDTPrediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_label(pred_LR[0]),
                                                                                                            output_label(pred_DT[0]),
                                                                                                            output_label(pred_GBC[0]),
                                                                                                            output_label(pred_RFC[0])))
```

```python
news = str(input())
manual_testing(news)
```
